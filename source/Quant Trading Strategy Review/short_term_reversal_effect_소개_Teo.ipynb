# 📈 단기 반전 전략(Short-Term Reversal Strategy)
# https://quantpedia.com/strategies/short-term-reversal-in-stocks/

 ### ※ 왜 주간 성과는 롱에, 월간 성과는 숏에 사용하는지 그 이론적 배경을 설명
행동 금융학적 근거:
1. 시간 프레임별 투자자 행동 차이
단기 (주간 5일): 과잉반응 (Overreaction)
##### 단기 투자자들의 특징:
- 뉴스에 즉각적이고 감정적으로 반응
- 공포와 탐욕에 의한 과도한 매매
- "팔고 보자" 또는 "일단 사자" 심리

##### 결과: 단기적 가격 왜곡 발생
좋은 뉴스 → 과도한 상승 → 빠른 조정
나쁜 뉴스 → 과도한 하락 → 빠른 반등 ← 이것을 노림!

중기 (월간 21일): 모멘텀 과열 (Momentum Overshoot)

##### 중기 투자자들의 특징:
- 트렌드 추종 (Trend Following)
- "상승하는 주식을 계속 사자" 심리
- 펀드매니저들의 상대 성과 추구

##### 결과: 지속적 상승 후 큰 조정
지속적 매수 → 과도한 상승 → 결국 큰 하락 ← 이것을 노림!


##### 단기 하락의 특징:
- ✅ 빠른 회복: 감정적 매도는 보통 1-2주 내 회복
- ✅ 높은 확률: 과도한 하락은 반등할 가능성이 높음
- ✅ 제한된 손실: 추가 하락 여지가 제한적

##### 만약 주간 상승을 숏으로 했다면?
- ❌ 모멘텀 지속: 단기 상승은 더 이어질 수 있음
- ❌ 큰 손실: 추가 상승으로 인한 무제한 손실 가능
- ❌ 타이밍 문제: 언제 조정될지 예측 어려움

##### 중기 상승의 특징:
- ✅ 조정 압력: 21일간 상승은 수익 실현 압력 높음
- ✅ 평균 회귀: 기본 가치로 돌아가려는 경향
- ✅ 리스크/보상: 조정 시 큰 하락 가능성

##### 만약 월간 하락을 롱으로 했다면?
- ❌ 추세 지속: 중기 하락은 더 이어질 수 있음  
- ❌ 가치 함정: 실제 기업 가치 하락일 수 있음
- ❌ 긴 회복: 중기 하락의 회복은 오래 걸릴 수 있음

### 경험적 연구 결과(Experimental Evidence)

많은 논문에서 단기(1~5일) 가격 반전 현상과 중기(1~12개월) 모멘텀 지속 현상이 함께 관찰

##### Jegadeesh (1990) 연구: "단기 반전 효과는 1주~1개월 구간에서 가장 강함"

##### De Bondt & Thaler (1985) 연구: "장기 반전 효과는 3~5년 구간에서 나타남"

##### Jegadeesh & Titman (1993) 연구: "중기 모멘텀 효과는 3~12개월 구간에서 강함"
